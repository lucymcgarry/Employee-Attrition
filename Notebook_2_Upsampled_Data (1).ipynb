# Modelling

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder

pipeline_categorical = Pipeline([
  ('onehot', OneHotEncoder(handle_unknown="ignore")),
])

pipeline_numerical = Pipeline([
  ('scaler', MinMaxScaler(feature_range=(0,1))),
])

pipeline_full = ColumnTransformer([
  ("categorical", pipeline_categorical, categorical_vars),
  ("numerical", pipeline_numerical, quantitative_vars),
])
```

```python
pipeline_full.fit(X_train)
X_train_transformed = pipeline_full.transform(X_train)
X_test_transformed = pipeline_full.transform(X_test)
```

```python
print(f"X_train_transformed.shape: {X_train_transformed.shape}")
```

```python
#K-Nearest Neighbours
#Decision Tree
#Random Forest
#SVC
#GradientBoost
#Logistic Regression
#Voting: Hard and Soft
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
```

```python
def evaluate_model(model, x_test, y_test):
    from sklearn import metrics

    # Predict Test Data
    y_pred = model.predict(x_test)

    # Calculate accuracy, precision, recall, f1-score, and kappa score
    acc = metrics.accuracy_score(y_test, y_pred)
    prec = metrics.precision_score(y_test, y_pred)
    rec = metrics.recall_score(y_test, y_pred)
    f1 = metrics.f1_score(y_test, y_pred)
    kappa = metrics.cohen_kappa_score(y_test, y_pred)

    # Calculate area under curve (AUC)
    fpr=0
    tpr=0
    auc=0
    try:
        y_pred_proba = model.predict_proba(x_test)[::,1]
        fpr, tpr, _ = metrics.roc_curve(y_test, y_pred_proba)
        auc = metrics.roc_auc_score(y_test, y_pred_proba)
    except:
        pass

    # Display confusion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)

    return {'acc': acc, 'prec': prec, 'rec': rec, 'f1': f1, 'kappa': kappa,
            'fpr': fpr, 'tpr': tpr, 'auc': auc, 'cm': cm}
```

```python
treeclf=DecisionTreeClassifier(random_state=42)
treeclf.fit(X_train_transformed, y_train)
```

```python
treeclf_eval = evaluate_model(treeclf, X_test_transformed, y_test)
print('Accuracy:', treeclf_eval['acc'])
print('Precision:', treeclf_eval['prec'])
print('Recall:', treeclf_eval['rec'])
print('F1 Score:', treeclf_eval['f1'])
print('Cohens Kappa Score:', treeclf_eval['kappa'])
print('Area Under Curve:', treeclf_eval['auc'])
print('Confusion Matrix:\n', treeclf_eval['cm'])
```

# Hyperparameter tuning

```python
import time
time_start = time.time()
from scipy.stats import randint, uniform, reciprocal, norm
from sklearn.model_selection import RandomizedSearchCV

rfclf_tuned=RandomForestClassifier(random_state=42)

hyperparameters = {
    'n_estimators': np.arange(50, 300, 10),#higher number of trees is good but can overfit at a certain point
    'max_depth': [None, np.arange(2, 10)],#higher depth leads to overfitting
    'min_samples_split': np.arange(2, 10), #to control overfitting - too high can cause under-fitting
    'min_samples_leaf': np.arange(1, 10), #lower values generally best in imbalanced data
    'max_features': {“sqrt”, “log2”, None}
    }

model = RandomizedSearchCV(rfclf_tuned, hyperparameters, n_iter=100, scoring="f1", cv=5, n_jobs=-1, verbose=10)
model.fit(X_train_transformed, y_train)
best_params = model.best_params_
print('Randomized Search CV done! Time elapsed: {} seconds'.format(time.time()-time_start))
print(best_params)
```
